# Scrap team statistics and conversion rates for each team over multiple years
## First Up - Team Statistics

```python
#Step 1: Create the URL for Team Stats

# Creates the list of seasons to scrap
seasons = [str(season) for season in range (2021, 2024)]
print(f'number of seasons = {len(seasons)}')

#Create the list of team abbreviations 
team_abbrs = ['atl','rav','buf','car','chi','cin','cle','dal',
              'den','det','gnb','htx','clt','jax','kan','sdg','ram',
              'rai','mia','min','nwe','nor','nyg','nyj','phi','pit','sea',
             'sfo','tam','oti','was','crd']
print(f'number of teams = {len(team_abbrs)}')

urls = []

for season in seasons:
    for team in team_abbrs:
        url = 'https://www.pro-football-reference.com/teams/'+team+'/'+season+'.htm'
        urls.append(url)

print(urls)
```

```python
#So now how would you do this for all 32 of the NFL teams?
#Create a pipeline
#Step 2: Define the table Scraping function

def table_scraper(urls):
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        table = soup.find('table')
        if table is None:
            print(f'No table found on {url}')
            return None
        headers = [th.text for th in table.find("thead").find_all("th")]
        headers = headers[12:]
        data = []
        for row in table.find("tbody").find_all("tr"):
            data.append([td.text for td in row.find_all("td")])
        df = pd.DataFrame(data, columns=headers)
        df = df.drop([2, 3])
        return df
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None
```

```python
#Step 3: Iterate over the set of web address
start = time.time()
Offensive_stats =[]
Defensive_stats =[]

for url in urls:
    table = table_scraper(url)
    time.sleep(random.randint(3,4)) #This is here to meet requirements from website - Max 20 inquires in a min
    print(url) #Not required but allows you to keep track of where you are in the process
    if table is not None:
        Offensive_stats.append(table.iloc[[0]])
        Defensive_stats.append(table.iloc[[1]])
end = time.time()
print(f"Execution time: {end - start:.6f} seconds")
#Runtime approx. 40mins = 2410sec - this was 32 teams over 20 years
```

```python
# NFL Standings website for bracket building 
# https://www.pro-football-reference.com/years/2023/ - pull AFC/NFC Playoff Standings

#Step 1: Create the URL Season Playoff Standings

# Creates the list of seasons to scrap
seasons = [str(season) for season in range (2021, 2024)]
print(f'number of seasons = {len(seasons)}')

urls = []

for season in seasons:
    url = 'https://www.pro-football-reference.com/years/'+season+ '/'
    urls.append(url)

print(urls)
```

```python
#Step 2: Define the table Scraping function

def table_scraper(urls):
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        table = soup.find('table', id = 'afc_playoff_standings')
        if table is None:
            print(f'No table found on {url}')
            return None
        headers = [th.text for th in table.find("thead").find_all("th")]
        headers = headers[:]
        data = []
        for row in table.find("tbody").find_all("tr"):
            data.append([td.text for td in row.find_all("td")])
        df = pd.DataFrame(data, columns=headers)
        df = df.drop([2, 3])
        return df
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return None
```

```python
def table_scraper_selenium(urls):
    # Set up the selenium WebDriver
    driver = webdriver.Chrome()
    driver.get(url)

    # Let the page load completely
    time.sleep(10)  # Wait for 10 seconds to ensure the page is fully loaded

    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Now find the table
    table = soup.find('table', id='afc_playoff_standings')
    driver.quit()  # Close the browser

    # Process the table as before
    if table:
        headers = [header.text for header in table.find_all('th')]
        #print("Headers found:", headers)

        # Extract additional headers starting from the 6th element onward
        extra_headers = headers[6:]
        #print("Extra headers:", extra_headers)

        data = []
        rows = table.find("tbody").find_all("tr")

        for i, row in enumerate(rows):
            cols = row.find_all("td")
            row_data = [td.text.strip() for td in cols]

            # Insert the corresponding extra header into the 'Tm' column
            if i < len(extra_headers):
                tm_data = extra_headers[i]
            else:
                tm_data = ""

            row_data.insert(0, tm_data)
            data.append(row_data)

        # Define the correct headers for the DataFrame
        correct_headers = ['Team', 'Wins', 'Losses', 'Ties', 'Seed', 'Reason']

        # Adjust row_data to match correct_headers length
        for i in range(len(data)):
            if len(data[i]) < len(correct_headers):
                data[i] = data[i] + [""] * (len(correct_headers) - len(data[i]))
            elif len(data[i]) > len(correct_headers):
                data[i] = data[i][:len(correct_headers)]

        # Create the DataFrame only if there is data
        if data:
            df = pd.DataFrame(data, columns=correct_headers)
            # Drop the last column of data
            df = df.iloc[:, :-1]
            # Keep only the first 6 rows
            df = df.head(7)
            # Update the 'Seed' column to be the index value + 1
            df['Seed'] = df.index + 1
            # Strip the (#) from the 'Team' column
            df['Team'] = df['Team'].str.replace(r'\s*\(\d+\)', '', regex=True)
            year = url.split('/')[-2]
            df.insert(0,'Season', year)
            print(df)
            return(df)
        else:
            print("No data found in table.")
    else:
        print('Table not found.')
```

```python
def table_scraper_selenium(url):
    # Extract the season year from the URL
    season = url.split('/')[-2]

    # Set up the selenium WebDriver
    driver = webdriver.Chrome()
    driver.get(url)

    # Let the page load completely
    time.sleep(10)  # Wait for 10 seconds to ensure the page is fully loaded

    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Now find the table
    table = soup.find('table', id='playoff_results')
    driver.quit()  # Close the browser

    # Process the table as before
    if table:
        headers = [header.text for header in table.find_all('th')]
        print("Headers found:", headers)

        # Extract additional headers starting from the 6th element onward
        extra_headers = headers[9:]
        print("Extra headers:", extra_headers)

        data = []
        rows = table.find("tbody").find_all("tr")

        for i, row in enumerate(rows):
            cols = row.find_all("td")
            row_data = [td.text.strip() for td in cols]

            # Insert the corresponding extra header into the 'Tm' column
            if i < len(extra_headers):
                tm_data = extra_headers[i]
            else:
                tm_data = ""

            row_data.insert(0, tm_data)
            data.append(row_data)

        # Define the correct headers for the DataFrame
        correct_headers = ['Round', 'Day', 'Date', 'Winner', 'Home / Away', 'Loser', 'Box Score', 'Winner Pts', 'Loser Pts']

        # Adjust row_data to match correct_headers length
        for i in range(len(data)):
            if len(data[i]) < len(correct_headers):
                data[i] = data[i] + [""] * (len(correct_headers) - len(data[i]))
            elif len(data[i]) > len(correct_headers):
                data[i] = data[i][:len(correct_headers)]

        # Create the DataFrame only if there is data
        if data:
            df = pd.DataFrame(data, columns=correct_headers)
            #Add column with the NFL Season
            df.insert(0, 'Season', season)
            #Drop Date and Box Score Columns
            #df.drop('Date','Box Score')
            print(df)
            return df
        else:
            print("No data found in table.")
            return None
    else:
        print('Table not found.')
        return None

# Example usage
seasons = [str(season) for season in range(2021, 2024)]
urls = ['https://www.pro-football-reference.com/years/' + season + '/' for season in seasons]

Playoff_Results = []



for url in urls:
    table = table_scraper_selenium(url)
    time.sleep(random.randint(3, 4))  # This is here to meet requirements from website - Max 20 inquires in a min
    print(url)  # Not required but allows you to keep track of where you are in the process
    if table is not None:
        Playoff_Results.append(table)
        Playoff_Results_df = pd.concat(Playoff_Results, axis = 0)

print(Playoff_Results_df)

print(f"Execution time: {end - start:.6f} seconds")
```

```python
# Get Things Setup that we will need 
Training_Playoff_df = []
Test_Playoff_df = []
Round_Results = []
Rounds = ['WildCard', 'Division', 'ConfChamp', 'SuperBowl']

# Define the unique seasons
Seasons = Playoff_Results_df['Season'].unique()

# Function to check if team data exists
def check_team_data(df, team, season):
    team_data = df[(df["Team"] == team) & (df["Season"] == season)]
    if team_data.empty:
        print(f"Data missing for {team} ({team_type}) in season {season}")
        return None
    return team_data.iloc[0].to_dict()

# Start looping through the Playoff_Results_df DataFrame 
for season in Seasons:
    Round_Results = Playoff_Results_df[(Playoff_Results_df['Season'] == season)]

    # Extract data for the current row
    for _, row in Round_Results.iterrows():
        Winner = row['Winner']
        Loser = row['Loser']
        Home_Game = row['Home']
        Neutral_Site = row['Neutral']
        Winner_Points = row['Winner Pts']
        Loser_Points = row['Loser Pts']
        Saturday_Game = row['Day_Sat']
        Sunday_Game = row['Day_Sun']
        Playoff_Round = row['Round']

        # Setup Winner and Loser Offensive and Defensive Stats - This needs to be filtered by team and season
        Winner_Offense = check_team_data(Combined_Offensive_Stats, Winner, season)
        Loser_Offense = check_team_data(Combined_Offensive_Stats, Loser, season)
        Winner_Defense = check_team_data(Combined_Defensive_Stats, Winner, season)
        Loser_Defense = check_team_data(Combined_Defensive_Stats, Loser, season)

        if Winner_Offense and Loser_Offense and Winner_Defense and Loser_Defense:
            # Combine Winner's Statlines
            Winner_Combined_Stats = {
                'Season': season,
                'Points Scored': Winner_Points,
                'Team': Winner,
                'Playoff_Round': Playoff_Round,
                'Saturday Game': Saturday_Game,
                'Sunday Game': Sunday_Game,
                'Home Game': Home_Game,
                'Neutral Site Game': Neutral_Site,
            }
            Winner_Combined_Stats.update({f"Off_{k}": v for k, v in Winner_Offense.items()})
            Winner_Combined_Stats.update({f"Def_{k}": v for k, v in Loser_Defense.items()})

            # Combine Loser's Statlines
            Loser_Combined_Stats = {
                'Season': season,
                'Points Scored': Loser_Points,
                'Team': Loser,
                'Playoff_Round': Playoff_Round,
                'Saturday Game': Saturday_Game,
                'Sunday Game': Sunday_Game,
                'Home Game': Home_Game,
                'Neutral Site Game': Neutral_Site,
            }
            Loser_Combined_Stats.update({f"Off_{k}": v for k, v in Loser_Offense.items()})
            Loser_Combined_Stats.update({f"Def_{k}": v for k, v in Winner_Defense.items()})

            # Create Training and Test Data Sets based on Season
            if int(season) < 2023:
                Training_Playoff_df.append(Winner_Combined_Stats)
                Training_Playoff_df.append(Loser_Combined_Stats)
            else:
                Test_Playoff_df.append(Winner_Combined_Stats)
                Test_Playoff_df.append(Loser_Combined_Stats)

# Convert Training and Test lists into Dataframes
Training_Playoff_df = pd.DataFrame(Training_Playoff_df)
Test_Playoff_df = pd.DataFrame(Test_Playoff_df)
```

```python
# Initialize lists for training and test data
training_playoff_df = []
test_playoff_df = []

# Function to combine stats
def combine_stats(row, offense, defense):
    combined_stats = {
        'Season': row['Season'],
        'Team': row['Team'],
        'Points Scored': row['Points Scored'],
        'Playoff_Round': row['Playoff_Round'],
        'Saturday Game': row['Saturday Game'],
        'Sunday Game': row['Sunday Game'],
        'Home Game': row['Home Game'],
        'Neutral Site Game': row['Neutral Site Game'],
    }
    combined_stats.update({f"Off_{k}": v for k, v in offense.items()})
    combined_stats.update({f"Def_{k}": v for k, v in defense.items()})
    return combined_stats

# Start looping through the Playoff_Results_df DataFrame 
for season in seasons:
    round_results = Playoff_Results_df[Playoff_Results_df['Season'] == season]

    for _, row in round_results.iterrows():
        winner = row['Winner']
        loser = row['Loser']
        winner_points = row['Winner Pts']
        loser_points = row['Loser Pts']
        home_game = row['Home']
        neutral_site = row['Neutral']
        saturday_game = row['Day_Sat']
        sunday_game = row['Day_Sun']
        playoff_round = row['Round']

        winner_offense = check_team_data(Combined_Offensive_Stats, winner, season, "Offense")
        loser_offense = check_team_data(Combined_Offensive_Stats, loser, season, "Offense")
        winner_defense = check_team_data(Combined_Defensive_Stats, winner, season, "Defense")
        loser_defense = check_team_data(Combined_Defensive_Stats, loser, season, "Defense")

        if winner_offense and loser_offense and winner_defense and loser_defense:
            winner_combined_stats = combine_stats({
                'Season': season,
                'Team': winner,
                'Points Scored': winner_points,
                'Playoff_Round': playoff_round,
                'Saturday Game': saturday_game,
                'Sunday Game': sunday_game,
                'Home Game': home_game,
                'Neutral Site Game': neutral_site,
            }, winner_offense, loser_defense)

            loser_combined_stats = combine_stats({
                'Season': season,
                'Team': loser,
                'Points Scored': loser_points,
                'Playoff_Round': playoff_round,
                'Saturday Game': saturday_game,
                'Sunday Game': sunday_game,
                'Home Game': home_game,
                'Neutral Site Game': neutral_site,
            }, loser_offense, winner_defense)

            # Append to appropriate list
            if int(season) < 2023:
                training_playoff_df.append(winner_combined_stats)
                training_playoff_df.append(loser_combined_stats)
            else:
                test_playoff_df.append(winner_combined_stats)
                test_playoff_df.append(loser_combined_stats)

# Convert lists to DataFrames
training_playoff_df = pd.DataFrame(training_playoff_df)
test_playoff_df = pd.DataFrame(test_playoff_df)

# Save to CSV files
training_playoff_df.to_csv(f'{file_path}\\Training_Playoff_Way2.csv', index=False)
test_playoff_df.to_csv(f'{file_path}\\Test_Playoff_Way2.csv', index=False)
```
